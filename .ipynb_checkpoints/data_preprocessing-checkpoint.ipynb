```python
amount = doc[["Transaction_ID", "Amount"]]
amount.isnull().sum() # 결측치 없음
amount.head()
amount.to_csv("amount.csv", index=False) # index제외하고 amount만 저장
```

```python
transaction_type = doc[["Transaction_ID", "Transaction_Type"]]
transaction_type.isnull().sum() # 결측치 없음
transaction_type.head()
transaction_type.to_csv("transaction_type.csv", index=False) # index제외하고 transaction_type만 저장
```

```python
system_latency = doc[["Transaction_ID", "System_Latency"]]
system_latency.isnull().sum() # 결측치 없음
system_latency.head()
system_latency.to_csv("system_latency.csv", index=False) # index제외하고 system_latency만 저장
```

```python
login_frequency = doc[["Transaction_ID", "Login_Frequency"]]
login_frequency.isnull().sum() # 결측치 없음
login_frequency.head()
login_frequency.to_csv("login_frequency.csv", index=False) # index제외하고 login_frequency만 저장
```

```python
failed_attempts = doc[["Transaction_ID", "Failed_Attempts"]]
failed_attempts.isnull().sum() # 결측치 없음
failed_attempts.head()
failed_attempts.to_csv("failed_attempts.csv", index=False) # index제외하고 failed_attempts만 저장
```

```python
payment_method = doc[["Transaction_ID", "Payment_Method"]]
payment_method.isnull().sum() # 결측치 없음
payment_method.head()
payment_method.to_csv("payment_method.csv", index=False) # index제외하고 failed_attempts만 저장
```

```python
category = doc[["Transaction_ID", "Category"]]
category.isnull().sum() # 결측치 없음
category.head()
category.to_csv("category.csv", index=False) # index제외하고 failed_attempts만 저장
```

```python
risk_incident = doc[["Transaction_ID", "Risk_Incident"]]
risk_incident.isnull().sum() # 결측치 없음
risk_incident.head()
risk_incident.to_csv("risk_incident.csv", index=False) # index제외하고 failed_attempts만 저장
```
